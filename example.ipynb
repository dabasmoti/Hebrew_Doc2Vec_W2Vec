```python
from gensim.models.doc2vec import Doc2Vec
import hebrew_tokenizer as ht
import pandas as pd
import glob
import platform
import os


model_path = r'C:\Users\Administrator\Dropbox\text\dbow_W3_EP15_yes_remove_stp_2020-02-13_1148\model/dbow_w3_EP15_yes_remove_stp'
model = Doc2Vec.load(model_path)
```

```python
test_articles_path = r'C:\Users\Administrator\Dropbox\text\dbow_W3_EP15_yes_remove_stp_2020-02-13_1148\model\today'
#test_articles_path = r'C:\Users\Administrator\Dropbox\text\NewDocs'
test_articles = glob.glob(test_articles_path+'\*.txt')
```

```python
def get_tokens(p):
    splt = '/' if platform.system() == 'Linux' else '\\'
    f_name = p.split(splt)[-1][:-4]
    with open(p,'r',encoding='utf-8') as f:
            return f_name,[token for grp, token, token_num, (start_index, end_index) in ht.tokenize(f.read()) if grp == 'HEB' ] 

def df_vectors(files):
    for file in file:
        file_name, tkn = get_tokens(file)
```

```python
articles_df = pd.DataFrame()
for file in test_articles:
    file_name, tkn = get_tokens(file)
    inferred_vector = model.infer_vector(tkn,steps=25)
    temp_df = pd.DataFrame(data=inferred_vector.reshape(1,300))
    temp_df.insert(loc=0,column ='article_id',value=file_name)
    articles_df = pd.concat([articles_df,temp_df])
articles_df.reset_index(inplace=True,drop=True)    
```
